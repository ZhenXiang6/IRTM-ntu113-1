```python
with open('./1.txt', 'r') as file:
    content = file.read()
```

```python
tokenization = content.replace('\"',' ').replace('-', ' ').replace('\'',' ').replace('?', ' ').replace(',', ' ').replace('.', ' ').replace('\n', ' ').split()
lowercased = [word.lower() for word in tokenization]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed = [ps.stem(x) for x in lowercased]
```

```python
word_freq = {}
for word in stemmed:
    if word not in word_freq:
        word_freq[word] = 1
    else:
        word_freq[word] += 1
```

```python
# show the frequency of each word in ascending order
import operator
sorted_word_freq = sorted(word_freq.items(), key=operator.itemgetter(1))
sorted_word_freq
```

```python
stopwords = ['the','to', 'a', 'of', 'in', 'it', 'is', 'will', 'on', 's', 'are']
without_stopwords = [word for word in stemmed if word not in stopwords]

with open('./out.txt', 'w') as file:
    for word in without_stopwords:
        file.write(word + '\n')
```
